```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
#Experience 1
#X_1 = angle alpha
#X_2 = eclairement

x = np.arange(0, 360, 10)
y_1 = np.asarray([70,50,10,0,0,0,0,0,20,50,70,90,120,130,130,130,110,90,70,40,20,0,0,0,0,0,20,50,80,100,120,130,130,130
                 ,120,90])

y_1 = y_1/max(y_1)

plt.plot(x, y_1, 'k+')
plt.xlabel("Angle")
plt.ylabel("Intensité lumineuse normalisée")
plt.savefig("laser.png")
```

```python
#experience 2 avec lumiere blanche et deux polariseur (le premier a 0)
y_2 = 10*np.array([110,109,102,88,68,50,32,17,7,1,0,8,22,40,58,77,94,106,111,108,101,87,69,48,30,13,5,1,
                  0,9,21,39,58,77,92,102])


y_2 = y_2 / max(y_2)
y_3 = np.cos(x*np.pi/180)**2


plt.plot(x, y_3, '-r')
plt.plot(x, y_2, 'k+')
plt.xlabel("Angle")
plt.ylabel("Intensité lumineuse normalisée")
plt.title(f"R² = {round(r2_score(y_2, y_3), 3)}")
plt.savefig("lampe.png")
```
